```python
## https://www.acmicpc.net/problem/1934

T = int(input())
for _ in range(T):
    A, B = map(int, input().split())

    head = A * B
    ## 유클리드 호제법. -> 두 수의 곱 / 두 수의 최대공약수
    while A % B != 0:
        A, B = B, A % B
    tail = B

    print(head // tail)

```

```python
## https://www.acmicpc.net/problem/1735

def gcd(A, B):
    while A % B != 0:
        A, B = B, A%B

    return B

h1, t1= map(int, input().split())
h2, t2 = map(int, input().split())

## 분수의 합.
## 분모가 서로 다른 경우 통분한다.
if t1 != t2:
    t3 = t1 * t2
    h1 *= (t3 // t1)
    h2 *= (t3 // t2)
else:
    t3 = t1

h3 = h1 + h2

## 기약 분수는 분모, 분자의 최대 공약수로 나눈 값.
res_h = h3 // gcd(h3, t3)
res_t = t3 // gcd(h3, t3)
print(res_h, res_t)
```

```python
## https://www.acmicpc.net/problem/4134
"""
정수론을 이용하는 문제. 양의 정수 M이 합성수(1과 자신외 다른 약수를 가진 양의 정수)인 경우
M은 두 개의 자연수이자 약수 a, b의 곱으로 표현될 수 있으므로 sqrt(M)보다 작거나 같은 약수를 반드시 가지고 있다.
"""
import sys

def is_prime(n):
    # for i in range(2, n // 2 + 1):
    for i in range(2, int(n ** 0.5)+1):
        if n % i == 0:
            return False ## 
    
    return True

n = int(sys.stdin.readline())

for _ in range(n):
    tmp = int(sys.stdin.readline())

    ## tmp와 같거나 큰 소수들 중 최소값.
    while True:
        if tmp == 0 or tmp == 1:
            print(2)
            break
            
        if is_prime(tmp):
            print(tmp)
            break
        else:
            tmp += 1
```

```python
## https://www.acmicpc.net/problem/17103

def eratosthenes_sieve(n):
    arr = [1] * (n+1)
    arr[0] = arr[1] = False
    for i in range(2, int(n**0.5) + 1):
        if arr[i]:
            j = 2
            while i * j <= n:
                arr[i * j] = 0
                j += 1

    return arr


N = int(input())
numbers = [int(input()) for _ in range(N)]
max_number = max(numbers)
prime_arr = eratosthenes_sieve(max_number)

for number in numbers:
    count = 0
    for i in range(2, number // 2 + 1):
        if prime_arr[i] and prime_arr[number - i]:
            count += 1
            
    print(count)
```
