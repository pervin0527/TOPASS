```python
## https://www.acmicpc.net/problem/10815
## 이진 탐색

N = int(input())
card_numbers = list(map(int, input().split()))
card_numbers.sort()

M = int(input())
target_numbers = list(map(int, input().split()))


card_numbers.sort()
for target in target_numbers:
    found = False
    start, end = 0, len(card_numbers) - 1
    while start <= end:
        mid = (start + end) // 2
        if card_numbers[mid] == target:
            print(1, end=" ")
            found = True
            break
        elif card_numbers[mid] > target:
            end = mid - 1
        else:
            start = mid + 1
    if not found:
        print(0, end=" ")
```

```python
## https://www.acmicpc.net/problem/7785
"""
입력과 동시에 처리하는 것보다 입력을 모두 받은 상태에서 처리하는 과정이 더 빠르다???
"""

check_list = {}
n = int(input())
for _ in range(n):
    name, log = input().split()
    check_list[name] = log

for name, log in list(check_list.items()):
    if log != "enter":
        del check_list[name]

res = sorted(list(check_list.keys()), reverse=True)
for r in res:
    print(r)
```

```python
## https://www.acmicpc.net/problem/1620
import sys

N, M = map(int, input().split())

ntoi = {}
for i in range(N):
    name = sys.stdin.readline().strip()
    ntoi.update({name : str(i+1)})
    ntoi.update({str(i+1) : name})

for _ in range(M):
    problem = sys.stdin.readline().strip()
    print(ntoi[problem])
```

```python
## https://www.acmicpc.net/problem/10816

import sys

N = int(input())
cards = list(map(int, sys.stdin.readline().strip().split()))

cards_dict = {}
for _ in range(N):
    if cards[-1] in cards_dict.keys():
        cards_dict[cards.pop()] += 1
    else:
        cards_dict[cards.pop()] = 1

M = int(input())
nums = list(map(int, sys.stdin.readline().strip().split()))
for num in nums:
    if num in cards_dict:
        print(cards_dict[num], end=" ")
    else:
        print(0, end=" ")
```

```python
## https://www.acmicpc.net/problem/1764

import sys

N, M = map(int, input().split())

names_dict = {}
for _ in range(N+M):
    name = sys.stdin.readline().rstrip()

    if name not in names_dict:
        names_dict[name] = 1
    else:
        names_dict[name] += 1

for name, freq in list(names_dict.items()):
    if freq != 2:
        del names_dict[name]

print(len(names_dict))
for name in list(names_dict.keys()):
    print(name)
```
