```python
## https://www.acmicpc.net/problem/2750

N = int(input())
arr = []

for _ in range(N):
    t = int(input())

    if len(arr) == 0:
        arr.append(t)
    else:
        for idx, n in enumerate(arr):
            if n > t:
                arr.insert(idx, t) ## idx - 1 이 아닌 것에 주의. 리스트의 길이가 계속해서 갱신(가변적).
                break
        else:
            arr.append(t)

    # print(arr)

for n in arr:
    print(n)
```

```python
## https://www.acmicpc.net/problem/2587
## 다섯 개의 자연수가 주어질 때 이들의 평균과 중앙값을 구하는 프로그램을 작성하시오.

arr = []
for _ in range(5):
    t = int(input())

    if len(arr) == 0:
        arr.append(t)
    else:
        for idx, n in enumerate(arr):
            if n > t:
                arr.insert(idx, t)
                break
        else:
            arr.append(t)

print(sum(arr) // 5)
print(arr[2])
```

```python
## https://www.acmicpc.net/problem/25305
## 주어진 리스트를 크기 순으로 정렬하는 방법???(빌트인 또는 클래스 메서드를 사용하지 않고)
N, k = map(int, input().split())
scores = list(map(int, input().split()))

for i in range(N):
    for j in range(i+1, N):
        if scores[j] > scores[i]:
            scores[i], scores[j] = scores[j], scores[i]

print(scores[k-1])
```

```python
## https://www.acmicpc.net/problem/10989

import sys

"""
일반적으로 수를 리스트에 추가하며 정렬하는 방식이 아니라, 입력될 수 있는 최대 가짓수 만큼의 리스트를 미리 할당하고, 0으로 초기화한다.
각각의 idx는 그 수를 의미하고, 값만큼 해당 idx를 출력한다.
"""

n = int(sys.stdin.readline())
check_list = [0] * 10001

for _ in range(n):
    check_list[int(sys.stdin.readline())] += 1

for i in range(10001):
    if check_list[i] != 0:
        for j in range(check_list[i]):
            print(i)
```

```python
## https://www.acmicpc.net/problem/1427

N = input()

check_list = [0] * 10
for c in N:
    check_list[int(c)] += 1

res = ""
for idx, item in enumerate(check_list):
    res += (str(idx) * item)

print(res[::-1])
```

```python
## https://www.acmicpc.net/problem/11650
N = int(input())

arr = []
for _ in range(N):
    x, y = map(int, input().split())

    if not arr:
        arr.append((x, y))

    else:
        for idx, item in enumerate(arr):
            if x != item[0] and item[0] > x:
                arr.insert(idx, (x, y))
                break
            elif x == item[0] and item[1] > y:
                arr.insert(idx, (x, y))
                break
        else:
            arr.append((x, y))

for item in arr:
    print(item[0], item[1])
```

```python
## https://www.acmicpc.net/problem/11651

import sys
N = int(input())

arr = []
for _ in range(N):
    x, y = map(int, sys.stdin.readline().split())
    arr.append((x, y))

arr.sort(key=lambda item : (item[1], item[0]))
for item in arr:
    print(item[0], item[1])
```

```python
## https://www.acmicpc.net/problem/1181

import sys

N = int(input())

arr = []
for i in range(N):
    vocab = sys.stdin.readline().rstrip()
    if vocab not in arr:
        arr.append(vocab)

arr.sort(key=lambda x : (len(x), x))

for item in arr:
    print(item)
```

```python
## https://www.acmicpc.net/problem/10814

N = int(input())

arr = []
for idx in range(N):
    age, name = input().split()
    arr.append((int(age), name, idx))

arr.sort(key=lambda x : (x[0], x[2]))
for item in arr:
    print(item[0], item[1])
```

```python
## https://www.acmicpc.net/problem/18870

N = int(input())
coords = list(map(int, input().split()))

sorted_coords = sorted(set(coords))
compress = {sorted_coords[i] : i for i in range(len(sorted_coords))}

for item in coords:
    print(compress[item], end=" ")
```
